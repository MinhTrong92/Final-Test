# Import và xử lýdữ liệu

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r'https://raw.githubusercontent.com/maxtran1997gmail/Files/main/dataset_2190_cholesterol.csv')
```

```python
df
```

**Kiểm tra dữ liệu có chứa giá trị null hay không?**

```python
df.info()
```

Dữ liệu không chứa giá trị null nhưng trong cột *ca* và *thal* có chứa giá trị "?"

```python
df[df.ca == "?"]
```

```python
df[df.thal == "?"]
```

**Loại bỏ giá trị đặc biệt trong các trường *ca* và *thal*. Thay thế giá trị này bằng mode của ca và thal**

```python
df = df.replace("?", np.nan)
```

```python
df.info()
```

```python
mode_ca = df['ca'].mode()[0]

mode_thal = df['thal'].mode()[0]
```

```python
df['ca'].fillna(mode_ca, inplace = True)

df['thal'].fillna(mode_thal, inplace = True)
```

```python
df
```

**Chuyển về định dạng số**

```python
df['ca'] = df['ca'].astype('float64')

df['thal'] = df['thal'].astype('float64')
```

```python
df.info()
```

# Phân chia các yếu tố đầu vào và đầu ra

```python
X = df.drop(['num'], axis=1)
```

```python
X
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
X_scaled
```

```python
y = df.num
```

```python
y
```

# Phân loại bằng Naive Bayes

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.5, random_state=0)
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

**Độ chính xác của mô hình khoảng 48.02%**

#  Phân loại bằng Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(solver='lbfgs', max_iter=1000)
lr = logreg.fit(X_train,y_train)
```

```python
y_pred=logreg.predict(X_test)
print(y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

**Độ chính xác của mô hình khoảng 58.55%**
